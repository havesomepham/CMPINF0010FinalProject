```python
%matplotlib inline
import pandas as pd
data_url = \
"https://data.wprdc.org/datastore/dump/37b2ac41-ae8e-4de1-8405-157e05dc3640"
chip = pd.read_csv(data_url)
chip.head()
```

```python
import matplotlib.pyplot as plt

# Assuming 'Category' is the column you want to use for the x-axis
category_counts = chip['neighborhood'].value_counts()

# Plot the bar graph
category_counts.plot(kind='bar', figsize=(10, 5))

# Set the title and labels
plt.title('Number of Speed Bumps in Each Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count')

# Show the plot
plt.show()
```

The top 5 neighborhoods for racing
1. Overbook
2. Stanton Heights
3. Central Oakland
4. Highland Park
5. Bloomfield
